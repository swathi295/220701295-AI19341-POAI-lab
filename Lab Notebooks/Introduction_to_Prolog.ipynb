```python
from pyswip import Prolog

# Initialize Prolog
prolog = Prolog()

# --- Facts ---
prolog.assertz("likes(john, lunch)")
prolog.assertz("likes(mary, party)")
prolog.assertz("likes(alex, air_guitar)")
prolog.assertz("likes(sara, party)")
prolog.assertz("likes(mike, air_guitar)")
prolog.assertz("attends(john, lunch)")
prolog.assertz("attends(mary, lunch)")
prolog.assertz("attends(alex, party)")
prolog.assertz("attends(sara, party)")
prolog.assertz("attends(mike, air_guitar)")

# --- Rules ---
prolog.assertz("foodie(X) :- likes(X, lunch), attends(X, lunch)")
prolog.assertz("party_person(X) :- likes(X, party), attends(X, party)")
prolog.assertz("musician(X) :- likes(X, air_guitar), attends(X, air_guitar)")

# Sample Queries
print("Foodies:")
for result in prolog.query("foodie(Person)"):
    print(result["Person"])

print("\nParty People:")
for result in prolog.query("party_person(Person)"):
    print(result["Person"])

print("\nMusicians:")
for result in prolog.query("musician(Person)"):
    print(result["Person"])
```
