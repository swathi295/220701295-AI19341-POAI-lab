```python
#Prolog

from pyswip import Prolog

# Initialize Prolog
prolog = Prolog()

# --- Facts ---
prolog.assertz("parent(john, mary)")
prolog.assertz("parent(john, mike)")
prolog.assertz("parent(susan, mary)")
prolog.assertz("parent(susan, mike)")
prolog.assertz("parent(mary, alice)")
prolog.assertz("parent(mary, bob)")
prolog.assertz("parent(mike, tom)")
prolog.assertz("male(john)")
prolog.assertz("male(mike)")
prolog.assertz("male(bob)")
prolog.assertz("male(tom)")
prolog.assertz("female(susan)")
prolog.assertz("female(mary)")
prolog.assertz("female(alice)")

# --- Rules ---
prolog.assertz("father(X, Y) :- parent(X, Y), male(X)")
prolog.assertz("mother(X, Y) :- parent(X, Y), female(X)")
prolog.assertz("sibling(X, Y) :- parent(Z, X), parent(Z, Y), X \\= Y")
prolog.assertz("grandparent(X, Y) :- parent(X, Z), parent(Z, Y)")
prolog.assertz("ancestor(X, Y) :- parent(X, Y)")
prolog.assertz("ancestor(X, Y) :- parent(X, Z), ancestor(Z, Y)")

# Sample Queries
print("Children of Mary:")
for result in prolog.query("parent(mary, Child)"):
    print(result["Child"])

print("\nIs John a grandparent of Alice?")
print("Yes" if list(prolog.query("grandparent(john, alice)")) else "No")

print("\nWho are Mary’s siblings?")
for result in prolog.query("sibling(mary, Sibling)"):
    print(result["Sibling"])

print("\nAncestors of Alice:")
for result in prolog.query("ancestor(Ancestor, alice)"):
    print(result["Ancestor"])
```
